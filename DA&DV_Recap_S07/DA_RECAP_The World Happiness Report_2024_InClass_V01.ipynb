### Context:

**[The World Happiness Report](https://worldhappiness.report/about/#:~:text=They%20include%20GDP%20per%20capita,freedom%2C%20generosity%2C%20and%20corruption.)** is a landmark survey of the state of global happiness. The World Happiness Report is a project of the United Nations Sustainable Development Solutions Network (SDSN). It is an annual publication that aims to assess and compare the happiness levels of countries worldwide. The first report was published in ``2012``, the second in ``2013``, the third in ``2015``, and the fourth in the ``2016`` Update. The World Happiness ``2017``, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th.

The World Happiness Report attempts to measure the happiness level of countries by using various factors. These factors include income levels, social support, life expectancy, freedom, generosity, corruption levels, and social support. The report ranks countries using the "Happiness Score" derived from the combination of these indicators.

The purpose of the report is to encourage policymakers and community leaders to develop policies that promote happiness. It also aims to understand the relationship between happiness and social, economic, and environmental factors by identifying and comparing happiness levels across different countries.

The World Happiness Report is created by analyzing various data sources. Data obtained from surveys, databases, and other sources are used to determine the happiness levels of countries. The report provides a ranking among countries and highlights specific factors that influence happiness levels.

The World Happiness Report is considered an important resource to guide governments in the policy-making process and enhance the well-being of societies. It has become a widely recognized reference worldwide and contributes to the development of effective policies that impact happiness.

### Content:

### What is Dystopia?
**[Dystopia](https://en.wikipedia.org/wiki/Dystopia)** is an imaginary country that has the worldâ€™s least-happy people. The purpose in establishing Dystopia is to have a benchmark against which all countries can be favorably compared (no country performs more poorly than Dystopia) in terms of each of the six key variables, thus allowing each sub-bar to be of positive width. The lowest scores observed for the six key variables, therefore, characterize Dystopia. Since life would be very unpleasant in a country with the worldâ€™s lowest incomes, lowest life expectancy, lowest generosity, most corruption, least freedom and least social support, it is referred to as â€œDystopia,â€ in contrast to Utopia. Additionally, dystopia was used as an antonym for **[Utopia](https://en.wikipedia.org/wiki/Utopia)** by **[John Stuart Mill](https://en.wikipedia.org/wiki/John_Stuart_Mill)** in one of his 1868 Parliamentary Speeches (Hansard Commons) by adding the prefix "dys" (Ancient Greek: Î´Ï…Ïƒ- "bad") to "topia"

```python
def first_looking(df):
    """
    Prints basic information about a column in a Pandas DataFrame.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to be analyzed.
        
    """
    
    display("shape_of_df    : ", df.shape)
    print("_______________________")
    display("num_of_nulls   : ", df.isnull().sum())
    print("_______________________")
    display("perc_of_nulls   : ", df.isnull().mean())
    print("_______________________")
    print("num_of_uniques : \n")
    for i in df.columns:
        print(f"{i}  : {df[i].astype(str).nunique():>}")  # >: Align the string to the right
                                                        
    print("_______________________")
    display("value_counts: ")
    for i in df.columns:
        print(f"column name: \n", df[i].value_counts(dropna=False))
        print("\n")
```

```python
def first_looking(df):
    """
    Prints basic information about a column in a Pandas DataFrame.
    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to be analyzed.
    """
    print(Fore.BLUE + "Shape of DataFrame", Fore.RESET)
    print(Fore.GREEN + "There have been " + Fore.RED + f"{df.shape[0]}" + Fore.GREEN + " rows and " + Fore.RED + f"{df.shape[1]}" + Fore.GREEN + " columns" + Fore.RESET)
    print("_"*40)
    print(Fore.BLUE + "Number of NULL values in the features\n", Fore.RESET)
    for i in df.columns:
        print(f"{i:<30}: {df[i].astype(str).nunique()}")
    print("_"*40)
    print(Fore.BLUE + "Percent of NULL values in the features\n", Fore.RESET)
    for column in df.columns:
        print(f"{column:<30}: {df[column].isnull().mean()}")
    print("_"*40)
    print(Fore.BLUE + "Number of UNIQUE values in the features\n", Fore.RESET)
    for i in df.columns:
        print(f"{i:<30}: {df[i].astype(str).nunique()}")
    print("_"*40)
    print(Fore.BLUE + "Value Counts:\n", Fore.RESET)
    for i in df.columns:
        print(Fore.GREEN + f"column name: {Fore.RED}{i}{Fore.RESET}")
        value_counts_df = df[i].value_counts(dropna=False).reset_index()
        aligned_value_counts_str = "\n".join([f"{index_val:<35}: {count}" for index_val, count in zip(value_counts_df.iloc[:, 0], value_counts_df.iloc[:, 1])])
        print(aligned_value_counts_str)
        print("\n")
    print("_"*40)
```

```python
first_looking(df)
```

```python
def first_looking_col(df, col):
    """
    Prints basic information about a column in a Pandas DataFrame.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to analyze.
    col : str
        The name of the column to analyze."""

    print("column name    : ", col)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df[col].isnull().mean(), 2))
    print("num_of_nulls   : ", df[col].isnull().sum())
    print("num_of_uniques : ", df[col].astype(str).nunique())
    print("lenght_of_the_column    : ", df.shape[0])
    print("--------------------------------")
    print("value_counts: \n", df[col].value_counts(dropna=False))
```

- If 'Healthy life expectancy' less than 75% of the 'Healthy life expectancy' score of the dataset but bigger than our threshold, it should be written **'Longer than normal expectancy'**

- If 'Healthy life expectancy' equal or less than our threshold but bigger than 25% of the 'Healthy life expectancy' score of the dataset, it should be written **'Normal life expectancy'**

- If 'Healthy life expectancy' equal or less than 25% of the 'Healthy life expectancy' score of the dataset, it should be written **'Lower life expectancy'**

### How to detect outliers applying Tukey's Fence Rule?

```python
# BONUS

def find_outliers(dataframe, column_name, threshold=1):
    # Create a copy of the dataframe
    dataframe_copy = dataframe.copy()
    
    # Get the values of the specified column
    column = dataframe_copy[column_name]
    
    # Calculate Quartile 1 (Q1) and Quartile 3 (Q3) values
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    
    # Calculate the Interquartile Range (IQR) value
    iqr = q3 - q1
    
    # Calculate the lower and upper thresholds to identify outliers
    lower_threshold = q1 - threshold * iqr
    upper_threshold = q3 + threshold * iqr
    
    # Create a mask to identify outliers
    outlier_mask = (column < lower_threshold) | (column > upper_threshold)
    
    # Select the rows with outliers
    outliers = dataframe_copy[outlier_mask]
    
    # Mark the outliers in a new column named "Outlier_Column"
    dataframe_copy.loc[outlier_mask, "Outlier_Column"] = column_name
    
    # Define the NaN values in "Outlier_Column" as "Not Outlier"
    dataframe_copy["Outlier_Column"].fillna("Not Outlier", inplace=True)
    
    return dataframe_copy, outliers
```

```python
# Find outliers
modified_df, outliers = find_outliers(df, 'logged_gdp_per_capita', threshold=1)

# Print the results
print("Modified DataFrame:")
display(modified_df)
print("\nOutliers:")
outliers
```

### How to detect outliers applying standart deviation?

```python
# BONUS

def find_outliers(dataframe, column_name, threshold=1):
    # Create a copy of the DataFrame
    dataframe_copy = dataframe.copy()
    
    # Get the values of the specified column
    column = dataframe_copy[column_name]
    
    # Calculate the mean and standard deviation of the column
    mean = column.mean()
    std = column.std()
    
    # Calculate the lower and upper thresholds to determine outliers
    lower_threshold = mean - threshold * std
    upper_threshold = mean + threshold * std
    
    # Create a mask to identify outliers
    outlier_mask = (column < lower_threshold) | (column > upper_threshold)
    
    # Select the rows with outliers
    outliers = dataframe_copy[outlier_mask]
    
    # Mark the outliers in a new column named "Outlier_Column"
    dataframe_copy.loc[outlier_mask, "Outlier_Column"] = column_name
    
    # Define the NaN values in "Outlier_Column" as "Not Outlier"
    dataframe_copy["Outlier_Column"].fillna("Not Outlier", inplace=True)
    
    return dataframe_copy, outliers
```

```python
# Find outliers
modified_df, outliers = find_outliers(df, 'logged_gdp_per_capita', threshold=2)

# Print the results
print("Modified DataFrame:")
display(modified_df)
print("\nOutliers:")
outliers
```

**ğŸ¤” Why do they give different results?**

The ``threshold=2 in Tukey's Fence Rule`` and the ``threshold=2 when using standard deviation`` do NOT have the same meaning. They use different approaches and can produce different results.

When using a standard deviation-based approach, the threshold=2 indicates that we consider a value as an outlier if it is more than 2 standard deviations away from the mean. This approach is based on the distribution of the data and defines outliers using the standard deviation.

On the other hand, Tukey's Fence Rule is based on the quartiles of the data. The threshold=2 determines the lower and upper threshold values by multiplying the IQR (Interquartile Range) value. If a value is smaller than Q1 - 2 * IQR or larger than Q3 + 2 * IQR, it is considered as an outlier. This method relies on the central tendency and distribution of the data.

Therefore, while the threshold=2 has a different meaning in a standard deviation-based approach, it has a different meaning in Tukey's Fence Rule. Both methods are different approaches to identifying outliers based on different statistical and distributional characteristics.

## How to detect Skewness & Kurtosis?

**Skewness and Kurtosis:**

[Skewness](https://www.investopedia.com/terms/k/kurtosis.asp) and [Kurtosis](https://www.investopedia.com/terms/s/skewness.asp) are two important statistical terms used in analyzing the distribution of data. They help us understand how the data is spread around the mean and how much it deviates from a normal distribution.

- **Skewness:**

  - Skewness measures the symmetry of the data distribution.
  - In a normal distribution, the skewness value is 0.
  - If the skewness value is positive, the distribution is right-skewed (with more data points above the mean).
  - If the skewness value is negative, the distribution is left-skewed (with more data points below the mean).

- **Kurtosis:**

  - Kurtosis measures the peakedness or flatness of the data distribution.
  - In a normal distribution, the kurtosis value is 3, which is Mesokurtic distribution used to describe the outlier characteristic of a probability distribution that is close to zero..
  - If the kurtosis value is greater than 3, the distribution is called "leptokurtic" and has heavier tails. This means there are more data points near the mean and more extreme data points.
  - If the kurtosis value is less than 3, the distribution is called "platykurtic" and is flatter than a normal distribution. This means there are fewer data points near the mean and fewer extreme data points.
